```python
%matplotlib inline
import matplotlib.pyplot as plt
from dotenv import load_dotenv
import os
import psycopg2
import yaml
import pandas as pd
import pandas.io.sql as sqlio
import ipyvolume as ipv
import ipyvolume.pylab as p3
import numpy as np
import ipywidgets
import folium
from keplergl import KeplerGl 
```

```python
with open(os.getcwd()+"/config.yml", 'r') as ymlfile:
    cfg = yaml.safe_load(ymlfile)
```

```python
PGHOST = cfg['db_creds']['PGHOST']
PGDATABASE =  cfg['db_creds']['PGDATABASE']
PGUSER =  cfg['db_creds']['PGUSER']
PGPASSWORD =  cfg['db_creds']['PGPASSWORD']
```

```python

def connect():
    
    # Set up a connection to the postgres server.
    conn_string = "host="+ PGHOST +" port="+ "5432" +" dbname="+ PGDATABASE +" user=" + PGUSER \
                  +" password="+ PGPASSWORD
    
    conn = psycopg2.connect(conn_string)
    print("Connected!")

    # Create a cursor object
    cursor = conn.cursor()
    
    return conn, cursor
```

```python
conn, cursor = connect()
```

```python
sql = "select * from public.data2_pre_processed_mv limit 50000;"
df = sqlio.read_sql_query(sql, conn)
```

```python
print(len(df))
```

```python
cursor.close()
conn.close()
```

```python
#keep the original to avoid doing another call to the db
df2 = df
```

```python
df2.head()
```

# Type of Incident recorded 

Incident type refers to M (Medical), F (Fire Dept) or P (Police)

```python
discipline_df = df2[['rep_id','discipline']].groupby('discipline').count()

group_names=discipline_df.index
group_size=discipline_df.rep_id


# Create colors
a, b, c=[plt.cm.Reds, plt.cm.Greens ,plt.cm.Blues]
 
# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _= ax.pie(group_size, radius=1.3, colors=[a(0.6), b(0.6), c(0.6)], labels= (group_size/len(df2))*100)
plt.setp( mypie, width=0.3, edgecolor='white')
plt.legend(group_names)
```

# Medical Incidents by Age-Group

```python
medical_incidents_df = df2[df2['discipline'] == 'M']
medical_incidents_df.to_csv('../processed-data/data2_medical_incidents.csv')
medical_incidents_grouped_df = medical_incidents_df[['age_group','rep_id']].groupby('age_group').count()

#print(medical_incidents_grouped_df )
group_names=medical_incidents_grouped_df.index
group_size=medical_incidents_grouped_df.rep_id


performance = group_size

plt.bar(group_names, group_size/len(medical_incidents_df), align='center', alpha=0.5)

plt.ylabel('Frequency')
plt.title('Medical Age Group Incidents')

plt.show()
```

# Top 10 Medical Incidents

```python
medical_incidents_grouped_df = medical_incidents_df[['chief_complaint_description','rep_id']].groupby('chief_complaint_description').count()

medical_incidents_grouped_df = medical_incidents_grouped_df.sort_values('rep_id', ascending=False)[0:10]
print(medical_incidents_grouped_df)
group_names=medical_incidents_grouped_df.index
group_size=medical_incidents_grouped_df.rep_id


performance = group_size
#fig= plt.figure(figsize=(15,7))
plt.bar(group_names, group_size/len(medical_incidents_df), align='center', alpha=0.5)
plt.xticks(rotation=45)
plt.ylabel('Frequency')
plt.title('Medical Incidents by Chief Complaint Description')

plt.show()
```

# TOP 5 Cities with Medical Incidents

```python
medical_incidents_grouped_df = medical_incidents_df[['ag_city','rep_id']].groupby('ag_city').count()

medical_incidents_grouped_df = medical_incidents_grouped_df.sort_values('rep_id', ascending=False)[0:5]
print(medical_incidents_grouped_df)
group_names=medical_incidents_grouped_df.index
group_size=medical_incidents_grouped_df.rep_id


performance = group_size
#fig= plt.figure(figsize=(15,7))
plt.bar(group_names, group_size/len(medical_incidents_df), align='center', alpha=0.5)
plt.xticks(rotation=45)
plt.ylabel('Frequency')
plt.title('Medical Incidents by City')

plt.show()
```

```python
fire_incidents_df = df2[df2['discipline'] == 'F']
fire_incidents_df.to_csv('../processed-data/data2_fire_incidents.csv')
fire_incidents_grouped_df = fire_incidents_df[['ag_city','rep_id']].groupby('ag_city').count()

fire_incidents_grouped_df = fire_incidents_grouped_df.sort_values('rep_id', ascending=False)[0:5]
print(fire_incidents_grouped_df)

group_names=fire_incidents_grouped_df.index
group_size=fire_incidents_grouped_df.rep_id


performance = group_size

plt.bar(group_names, group_size/len(fire_incidents_df), align='center', alpha=0.5)
plt.xticks(rotation=45)
plt.ylabel('Frequency')
plt.title('Fires by Cities')

plt.show()
```

```python
police_dates = pd.DataFrame(pd.to_datetime(police_incidents_df['start_time'].str.split('T').str[0].values), columns=['date'])
police_dates["_id"] = pd.to_datetime(police_dates["date"])
police_dates.index = pd.DatetimeIndex(police_dates["_id"])
police_incidents_by_day_df = police_dates.groupby(pd.Grouper(key='date',freq='D')).count()
police_incidents_by_day_df
plt.plot_date(police_incidents_by_day_df.index,police_incidents_by_day_df["_id"], fmt='-')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.title('Crimes by day')

plt.show()
```

```python

police_incidents_by_week_df = police_dates.groupby(pd.Grouper(key='date',freq='W')).count()
police_incidents_by_week_df
plt.plot_date(police_incidents_by_week_df.index,police_incidents_by_week_df["_id"], fmt='-')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.title('Crimes by week')

plt.show()
```

```python
police_incidents_by_month_df = police_dates.groupby(pd.Grouper(key='date',freq='M')).count()
police_incidents_by_month_df
plt.plot_date(police_incidents_by_month_df.index,police_incidents_by_month_df["_id"], fmt='-')
plt.xticks(rotation=45)
plt.ylabel('Count')
plt.title('Crimes by month')

plt.show()
```
